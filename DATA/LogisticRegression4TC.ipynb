```python
imp =pd.read_csv('imp.csv', delimiter = ',', header = 0)
```

```python
imp = imp.dropna(subset=['Pole_Rich'])
```

```python
raw = imp
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder().fit_transform(imp['Pole_Rich'])
le
```

```python
raw.loc[:, "Pole_Rich_Code"] = le
raw.head()
```

```python
raw.Pole_Rich_Code.value_counts()
```

```python
raw = raw.dropna(subset=['Pole_Rich_Code'])
```

```python
train, test = statsmodels_train_test_split(raw)
```

```python
formula = 'Pole_Rich_Code ~  Estimated_Number_of_Businesses + Total_Population + Population_Growth+ Utility_Pole_Make_Ready_Estimated_Required_Number_of_Moves'
```

```python
model = smf.ols(formula = formula, data=train).fit()
model.summary()
```

```python
y_pred_test = model.predict(test)
y_pred_test
```

```python
y_pred_test=y_pred_test.dropna()
```

```python
r2_score(test['Pole_Rich_Code'], y_pred_test)
```

```python
predictions = np.round(y_pred_test)
```

```python
confusion_matrix(test['Pole_Rich_Code'], predictions)
```
